```python
import pandas as pd
import matplotlib.pyplot as plt
```

выводим датасет

```python
genshin = pd.read_csv(r"C:\Users\User\Downloads\archive (1)\Genshin_Impact_All_Character_Stat.csv")

genshin
```

Линейный график базового значения атаки (Base ATK) для персонажей различных редкостей (Rarity)

```python
characters = genshin["Character"].unique()[:3]

for character in characters:
    character_data = genshin[genshin["Character"] == character]
    plt.plot(character_data["Base ATK"], character_data["Rarity"], label=character)

plt.xlabel("Base ATK")
plt.ylabel("Rarity")
plt.legend()
plt.show()
```

на графике отображается изменение базового значения атаки персонажей различных редкостей (4 звезды) с ростом уровня. видим, что с увеличением уровня атака персонажей также повышается.

Столбчатая диаграмма базового значения элемента (Element) для персонажей с разными основными ролями (Main role)

```python
characters = genshin["Character"].unique()[:5]

for character in characters:
    character_data = genshin[genshin["Character"] == character]
    plt.bar(character_data["Element"], character_data["Main role"], label=character, alpha=0.5)

plt.xlabel("Element")
plt.ylabel("Main role")
plt.legend()
plt.show()
```

На данном графике отображаются элементы персонажей различной основной роли.

  круговая диаграмма базового значения здоровья (Base HP) для персонажей разных элементов (Element)

```python
Element = genshin["Element"].unique()
Base_HP = genshin.groupby("Element")["Base HP"].sum()

plt.pie(Base_HP, labels=Base_HP, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

получилась диаграмма, показывающая хп персонажей и их процентное сообщение относительно элементов

круговая диаграмма оружия (Weapon) для персонажей разных редкостей (Rarity)

```python
Weapon = genshin["Weapon"].unique()
Rarity = genshin.groupby("Weapon")["Rarity"].sum()

plt.pie(Rarity, labels=Rarity, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

получилась еще одна круговая диаграмма, с данными оружия и редкости персонажей

далее создаем диаграму рассеяния

```python
characters = genshin["Character"].unique()[:5]

for character in characters:
    character_data = genshin[genshin["Character"] == character]
    plt.scatter(character_data["Weapon"], character_data["Element"], label=character)

plt.xlabel("Weapon")
plt.ylabel("Element")
plt.legend()
plt.show()
```

вывелся график рассеяния, демонстрирующий персонажей по элементу и типу оружия

создаем еще одну диаграмму рассеяния

```python
characters = genshin["Character"].unique()[:5]

for character in characters:
    character_data = genshin[genshin["Character"] == character]
    plt.scatter(character_data["Rarity"], character_data["Lv"], label=character)

plt.xlabel("Rarity")
plt.ylabel("Lv")
plt.legend()
plt.show()
```

получилась еще одна диаграмма рассеяния, показывающая персонажей по уровню и редкости

создаем гистограмму

```python
characters = genshin["Character"].unique()[:3]

for character in characters:
    character_data = genshin[genshin["Character"] == character]
    plt.hist(character_data["Lv"], label=character, alpha=0.5)

plt.xlabel("Lv")
plt.ylabel("Rarity")
plt.legend()
plt.show()
```

вывелась гистограмма, показывающая уровень и редкость персонажей

создаем еще одну гистограмму

```python
characters = genshin["Character"].unique()[:5]

for character in characters:
    character_data = genshin[genshin["Character"] == character]
    plt.hist(character_data["Main role"], label=character, alpha=0.5)

plt.xlabel("Main role")
plt.ylabel("Base ATK")
plt.legend()
plt.show()
```

вывелась гистограмма, показывающая роль персонажа и еще базовую атаку

создаем последнюю третью гистограмму

```python
characters = genshin["Character"].unique()[:5]

for character in characters:
    character_data = genshin[genshin["Character"] == character]
    plt.hist(character_data["Weapon"], label=character, alpha=0.5)

plt.xlabel("Weapon")
plt.ylabel("Base HP")
plt.legend()
plt.show()
```

вывелась последняя гистограмма, показывающая оружие и базовое ХП персонажей
